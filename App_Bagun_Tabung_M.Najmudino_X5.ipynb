```python
# program menghitung volume tabung

# meminta input dari pengguna
r = float(input("Masukkan jari-jari alas (cm): ") )
t = float(input("Masukkan tinggi tabung (cm): "))
pi = 3.14

# Menghitung volume
Volume = pi * r**2 * t

# Menampilkan hasila
print(f"volume tabung adalah: {Volume} cm^3")
```
