Aritmatika

```python
#Operand tanpa disimpan dalam variabel
print(5000*3)
```

```python
#Simpan operand ke dalam variabel
harga_kojek = 6000
banyak_porsi = 3

print(harga_kojek*banyak_porsi)
```

```python
#Simpan hasil hasil opreasi kedalam variabel
harga_total_kojek = 3000*3

print(harga_total_kojek)
```

```python
#Simpan nilai operand dan hasil operasi ke dalam variabel
harga_cilok = 1000
jumlah_porsi = 5
harga_total_cilok = harga_cilok*jumlah_porsi

print(harga_total_cilok)
```

```python
uang_yang_diterima = 100000
uang_kembalian = uang_yang_diterima - harga_total_cilok

print(uang_kembalian)
```

Jenis-jenis operator aritmatika pada python

```python
#Penjumlahan
iphone_rolan = 21
iphone_gibran = 4

total_iphone = iphone_rolan + iphone_gibran

print(total_iphone)
```

```python
#Pengurangan
iphone_rolan = 21
iphone_gibran = 4

selisih_iphone = iphone_rolan - iphone_gibran

print(selisih_iphone)
```

```python
#perkalian
jumlah_dus = 5
jumlah_botol_per_dus = 10

total_botol = jumlah_dus * jumlah_botol_per_dus

print(total_botol)
```

```python
#Pembagian
iphone_rolan = 21
jumlah_anak_yatim = 5

pembagian_iphone = iphone_rolan / iphone_gibran

print(pembagian_iphone)
```

```python
#Pembagian bulat
iphone_rolan = 21
jumlah_anak_yatim = 5

pembagian_iphone = iphone_rolan // iphone_gibran

print(pembagian_iphone)
```

```python
#sisa bagi

iphone_rolan = 21
jumlah_anak_yatim = 5

sisa_bagi_iphone = iphone_rolan % jumlah_anak_yatim

print(sisa_bagi_iphone)
```
